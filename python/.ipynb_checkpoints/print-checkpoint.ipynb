## PRINT - print()

```python
# First customary program "Hello, world!"
print("Hello, world!")
```

```python
# New Line
print("Hello\n world!")
```

```python
# No Space
print("Hello," + "world!")
```

```python
# Added Space
print("Hello, " + "world!")
```

```python
# Added Space
print("Hello," + " " + "world!")
```

```python
# Print Number
print(10)
```

```python
# Print Sum of Two Numbers
print(10 + 5)
```

```python
# Print "Hello " three times
print("Hello " * 3)
```
